```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ConstantKernel, Matern
```

```python
filename = "D:\IITG Onedrive\OneDrive - Indian Institute of Technology Guwahati\IITG\Beyond Course\TIM\Datasets\WK100_1.txt"
weights = []

with open(filename, 'r') as file:
    for line in file:
        words = line.split()
        if len(words) >= 3:
            weight = (words[2])
            weights.append(weight)

for i in range(len(weights)):
    if len(weights[i]) == 3:
        weights[i] = 1
    else:
        weights[i] = -1
```

```python
density = 50

for i in range(len(weights)):
    if np.random.randint(0,10) > density/10:
        weights[i] = 0
    else:
        weights[i] = abs(weights[i])
```

```python
nodes = [100, 120, 140, 160, 180, 200, 220, 240, 320, 380, 400, 420, 440, 500, 600, 700, 800]
times = [23.42, 29.45, 43.34, 55.01, 69.22, 90.07, 103.14, 123.31, 217, 310.5, 338.04, 379, 423, 
         460, 647, 878, 1233]
energy = [-383,-473,-628,-728,-852,-1052,-1193,-1313,-2069,-2696,
          -2829,-2993,-3089,-3629,-4369,-5525,-5885]
cut_values = [1435,2050,2788,3539,4394,5546,6591,7904,13762,
              19464,21305,23559,25533,32719,47053,64003,82983]

N = len(nodes)
time_fit = np.zeros(N)
energy_fit = np.zeros(N)
cut_fit = np.zeros(N)

for i in range(N):
    time_fit[i] = 0.0017*nodes[i]**2 + 0.1164*nodes[i]
    energy_fit[i] = -8.3321*nodes[i] + 565.9
    cut_fit[i] = 0.1269*nodes[i]**2 + 2.3781*nodes[i]
```

```python
%matplotlib notebook
plt.title("Density = 50")
plt.ylabel("Time taken(s)")
plt.xlabel("Nodes")
plt.plot(nodes, times, ' x', label='times')
plt.plot(nodes, time_fit, label='fit')
plt.legend()
plt.show()
```

```python
%matplotlib notebook
plt.xlabel("Nodes")
plt.ylabel("Energy")
plt.plot(nodes, energy, ' x', label='energy')
plt.plot(nodes, energy_fit, label='fit')
plt.legend()
```

```python
%matplotlib notebook
plt.xlabel("Nodes")
plt.ylabel("Best value of cut")
plt.plot(nodes, cut_values, ' x', label='Cut values')
plt.plot(nodes, cut_fit, label='fit')
```
